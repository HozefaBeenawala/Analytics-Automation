## 1. Which location has most sales ?

```python
df_Accepted = df[df['Order Status'] == "Accepted"]
```

```python
df_Accepted
```

```python
df_Accepted['Sales'] = (df_Accepted['Sales Rate']*df_Accepted['Quantity'])-df_Accepted['Discount']
```

```python
# Sort the DataFrame by 'Sales' column in descending order
df_sorted = df_Accepted.sort_values(by='Sales', ascending=False)

# Create a figure
fig = plt.figure(figsize=(20, 5))

# Create the bar plot using the sorted DataFrame
plt.bar(df_sorted['Location'], df_sorted['Sales'], color='maroon', width=0.4)

# Set labels and title
plt.xlabel("Location")
plt.ylabel("Sales")
plt.title("Location with Most Sales")

# Show the plot
plt.show()
```

## 2. Which Product has most sales ?

```python
# Sort the DataFrame by 'Sales' column in descending order
df_sorted = df_Accepted.sort_values(by='Sales', ascending=False)

# Create a figure
fig = plt.figure(figsize=(20, 5))

# Create the bar plot using the sorted DataFrame
plt.bar(df_sorted['Product name'], df_sorted['Sales'], color='maroon', width=0.4)

# Set labels and title
plt.xlabel("Product")
plt.ylabel("Sales")
plt.title("Product with Most Sales")

# Show the plot
plt.show()
```

```python
plt.pie(df_sorted['Sales'], labels = df_sorted['Product name'])
plt.show() 
```

```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
class Analysis:
    
    def __init__(self,data,sheet):
        self.data = data
        self.sheet = sheet
        self.filtered_df = None
        self.df = None
        
    def dataset(self):
        self.df = pd.read_excel(self.data,sheet_name=self.sheet)
        return self.df
    
    def join(self,data2,sheet2,column,method):
        df = pd.read_excel(data2,sheet_name=sheet2)
        self.df = pd.merge(self.df,df,on=column, how=method)
        return self.df
    
    def filter_data(self,column,value):
        self.filtered_df = self.df[self.df[column]==value]
        return self.filtered_df
    
    def nofilter_data(self):
        self.filtered_df = self.df
        return self.filtered_df
    
    def sales_column (self,name,rate_column,quantity_column,discount_column):
        self.df['Sales'] = (self.df[rate_column]*self.df[quantity_column])-self.df[discount_column]
        return self.df
    
    def location(self,sales_column,location_column):
        #df_Accepted = df[df[status_column] == status_value]
        #self.filtered_df[sales_column] = (self.filtered_df[rate_column]*self.filtered_df[quantity_column])-self.filtered_df[discount_column]
        # Sort the DataFrame by 'Sales' column in descending order
        df_sorted = self.filtered_df.sort_values(by=sales_column, ascending=False)
        
        # Create a figure
        fig = plt.figure(figsize=(20, 5))

        # Create the bar plot using the sorted DataFrame
        plt.bar(df_sorted[location_column], df_sorted[sales_column], color='maroon', width=0.4)
        
        # Set labels and title
        plt.xlabel("Location")
        plt.ylabel("Sales")
        plt.title("Location with Most Sales")

        # Show the plot
        plt.show()
        
    def product(self,sales_column,product_column):
        # Sort the DataFrame by 'Sales' column in descending order
        df_sorted = self.filtered_df.sort_values(by=sales_column, ascending=False)
        # Create a figure
        fig = plt.figure(figsize=(20, 5))
        plt.pie(df_sorted[sales_column], labels = df_sorted[product_column])
        plt.show()
    
        
        
    
    
    
    
    
    
        

        
```
