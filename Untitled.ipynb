```python
df_products = pd.read_excel('E-Commerce Sales Dataset.xlsx',sheet_name='Products')
```

```python
df_locations = pd.read_excel('E-Commerce Sales Dataset.xlsx',sheet_name='Location')
```

```python
df_oc = pd.merge(df_orders,df_customers,on="Customer ID",how="inner")
```

```python
df_oc
```

```python
df_ocp = pd.merge(df_oc,df_products,on="Product ID",how="inner")
```

```python
df_ocp
```

```python
df = pd.merge(df_ocp,df_locations,on="Location ID",how="inner")
```

```python
df
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

## 1. Which location has most sales ?

```python
df_Accepted = df[df['Order Status'] == "Accepted"]
```

```python
df_Accepted
```

```python
df_Accepted['Sales'] = (df_Accepted['Sales Rate']*df_Accepted['Quantity'])-df_Accepted['Discount']
```

```python
# Sort the DataFrame by 'Sales' column in descending order
df_sorted = df_Accepted.sort_values(by='Sales', ascending=False)

# Create a figure
fig = plt.figure(figsize=(20, 5))

# Create the bar plot using the sorted DataFrame
plt.bar(df_sorted['Location'], df_sorted['Sales'], color='maroon', width=0.4)

# Set labels and title
plt.xlabel("Location")
plt.ylabel("Sales")
plt.title("Location with Most Sales")

# Show the plot
plt.show()
```

## 2. Which Product has most sales ?

```python
# Sort the DataFrame by 'Sales' column in descending order
df_sorted = df_Accepted.sort_values(by='Sales', ascending=False)

# Create a figure
fig = plt.figure(figsize=(20, 5))

# Create the bar plot using the sorted DataFrame
plt.bar(df_sorted['Product name'], df_sorted['Sales'], color='maroon', width=0.4)

# Set labels and title
plt.xlabel("Product")
plt.ylabel("Sales")
plt.title("Product with Most Sales")

# Show the plot
plt.show()
```

## 3. Year wise sales

```python
df_Accepted['Order Date'] = df_Accepted['Order Date'].astype(str)
```

```python
df_Accepted['Year'] = df_Accepted['Order Date'].str.split('-').str[0]
df_Accepted['Month'] = df_Accepted['Order Date'].str.split('-').str[1]
df_Accepted['Date'] = df_Accepted['Order Date'].str.split('-').str[2]
```

```python
# Sort the DataFrame by 'Sales' column in descending order
df_sorted = df_Accepted.sort_values(by='Sales', ascending=False)

# Create a figure
fig = plt.figure(figsize=(20, 5))

# Create the bar plot using the sorted DataFrame
plt.bar(df_sorted['Year'], df_sorted['Sales'], color='maroon', width=0.4)

# Set labels and title
plt.xlabel("Year")
plt.ylabel("Sales")
plt.title("Year with Most Sales")

# Show the plot
plt.show()
```
